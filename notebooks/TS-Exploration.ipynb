```python

# Convert the 'time' column to datetime
first_day_df['time'] = pd.to_datetime(df['time'], format='%H:%M:%S')

# Create an empty list to store the rows for the final dataframe
rows = []

# Iterate through each row
for idx, row in df.iterrows():
    # Get the original time for this row
    base_time = pd.to_datetime(row['time'], format='%H:%M:%S' )
    
    # Iterate through the bg columns
    for col in df.columns:
        if col.startswith('bg'):
            # Extract the time delta from the column name
            time_str = col.split('-')[-1] if '-' in col else col.split('+')[-1]
            time_parts = time_str.split(':')
            
            # Convert hours and minutes from the column name to timedelta
            hours, minutes = int(time_parts[0]), int(time_parts[1])
            time_delta = timedelta(hours=hours, minutes=minutes)
            
            # Adjust the base time according to whether it's a past ('-') or future ('+') event
            if '-' in col:
                event_time = base_time - time_delta
            else:
                event_time = base_time + time_delta
            
            # Get the value of bg for this time
            bg_value = row[col]
            
            # Append the new row if bg_value is not NaN
            if not pd.isna(bg_value):
                rows.append({'time': event_time.time(), 'bg': bg_value})

# Create a new dataframe from the rows list
final_df = pd.DataFrame(rows)

# Sort the final dataframe by time
final_df.sort_values(by='time', inplace=True)
print(final_df)

```
